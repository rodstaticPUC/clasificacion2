# Vinos: Clasificacion

3 clases objetivo, 178 obs y 13 variables

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
```

```python
wine = load_wine()

x = range(0,wine.data.shape[0], 1)

scaler = MinMaxScaler()
scaler.fit(wine.data)
wine_scaled = scaler.transform(wine.data)
```

```python
def plot(init_feature, end_feature,mark): # decidimos en inicio y el final de la caracteristica
    plt.figure(figsize=(30,5))
    for i in range(init_feature, end_feature,1):
        plt.plot(x,wine_scaled[:,i],mark,label=wine.feature_names[i])
        plt.rcParams.update({'font.size':18});
        plt.legend(prop={'size':22});
        plt.grid();
        plt.show()
```

```python
plot(0,4,'-')
plot(4,8,'-')
plot(8,12,'')
```
